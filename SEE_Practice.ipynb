#### 1.  ii) Write a Python program to generate the first N Fibonacci numbers using both recursion and iteration. Compare their time complexity

```python
# Recursive method

def fib(n):
    if n<0:
        raise ValueError("Fibonacci is not defined in negative numbers")
    elif n <=1 :
        return n
    else:
        return fib(n-1) + fib(n-2)
while True:
    try:
        x = int(input("Enter a positive number to print fibonacci sequence: "))
        if x<0:
            raise ValueError
        break
    except ValueError:
        print("Invalid Input! Please enter a positive number..")
print(f"The first {x} Fibonacci numbers are: ")
for i in range (x):
    print(fib(i),end=" ")
```

```python
# iterative method
while True:
    try:
        n = int(input("Enter a number to print fibonacci sequence: "))
        if n<0:
            raise ValueError
        break
    except ValueError:
        print("Invalid Input! Please enter a positive number..")
a,b = 0,1
for i in range(n):
    print(a,end=" ")
    a,b = b,a+b
```

#### 2. i) Write a Python decorator that prints "Before calling function" and"After calling function" around the execution of any function.

```python
def decor(func):
    def wrapper():
        print("Before calling function ")
        func()
        print("After calling function")
    return wrapper

@decor
def name():
    print("Hello! from inside function")
name()
```

#### 2. ii) Write a decorator that converts the output of a function to uppercase.

```python
def upper_case(func):
    def wrapper(*args,**kwargs):
        result = func(*args,**kwargs)
        return result.upper()
    return wrapper

@upper_case
def greet(name):
    return f"{name}"
x = input("Enter any string: ")
print(greet(x))
```

#### 3 i) Write a Python program demonstrating encapsulation using a class BankAccount with private attributes __balance and methods to deposit and withdraw money

```python
class BankAccount:
    def __init__ (self,username,pin,initial_balance=0):
        self.username = username
        self.__pin = pin
        self.__balance = initial_balance
        
    def verifypin(self,enterdpin):
        return self.__pin == enterdpin
    
    def deposit(self,amount):
        if amount < 0:
            raise ValueError("Cannot deposit Negative amount!")
        self.__balance += amount
        print(f"{amount} depositted . Current Balance: ${self.__balance}")
        
    def withdraw(self,amount):
        if amount < 0:
            raise ValueError("Cannot withdraw Negative amount!")
        if amount > self.__balance:
            print("Insufficient Balance")
        else:
            self.__balance -= amount
            print(f"{amount} withdrawn. Current Balance: ${self.__balance}")
        
        
    
    def get_balance(self):
        return self.__balance
    
print("\t Welcome to our Secure Banking System \t")

username = input("Enter your name: ")
pin = int(input("Set a 4-digit PIN: "))
initial = float(input("Enter opening balance: $"))
account = BankAccount(username,pin,initial)

while True:
    print(f"\nHello {account.username} choose an operation...")
    print("1. Deposit \n2. Withdraw \n3. Current Balance \n4. Exit")
    
    choice = int(input("Enter your choice(1-4): "))
    
    try:
        if choice == 1:
            pin = int(input("Enter your PIN to proceed: "))
            if account.verifypin(pin):
                amount = float(input("Enter the amount to deposit: "))
                account.deposit(amount)
            else:
                print("Invalid PIN")
        
        elif choice == 2:
            pin = int(input("Enter your PIN to proceed: "))
            if account.verifypin(pin):
                amount = float(input("Enter amount to withdraw: "))
                account.withdraw(amount)
            else:
                print("Invalid PIN")
                
        elif choice == 3:
            print(f"Current Balance: {account.get_balance()}")
            
        elif choice == 4:
            print("\nThank you for Banking with us! ")
            break
            
        else:
            print("Invalid choice! Please enter 1-4")
            
    except ValueError as Ve:
        print(Ve)
```

#### 3. ii) Write a Python program that demonstrates abstraction using an interface (abstract class) for different payment methods like CreditCard, PayPal, and Bitcoin

```python
from abc import ABC, abstractmethod

class paymentmethod(ABC):
    @abstractmethod
    def pay(self,amount):
        pass
    
class creditcard(paymentmethod):
    def pay(self,amount):
        print(f"Paid {amount} using creditcard")
        
class paypal(paymentmethod):
    def pay(self,amount):
        print(f"Paid {amount} using PayPal")

class Bitcoin(paymentmethod):
    def pay(self,amount):
        print(f"Paid {amount} using Bitcoin")
        
print("Welcome Payment Gateway")
print("Choose a payment method")
print("1. Creditcard \n2. PayPal \n3. Bitcoin")

choice = int(input("Enter your choice (1-3): "))
amount = float(input("Enter the amount to pay: "))

if choice == 1:
    method = creditcard()

elif choice == 2:
    method = paypal()
    
elif choice == 3:
    method = Bitcoin()

else:
    print("Invalid Choice")
    method = None
    
if method:
    method.pay(amount)
```
